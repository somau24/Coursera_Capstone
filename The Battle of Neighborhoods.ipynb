<h1>The Battle of Neighborhoods | Introduction/Business Problem</h1>

<h2> A description of the problem and a discussion of the background</h2>

we are a consulting company helping the small and medium sized enterprises (SMEs) to enter the market and develop themselves on the Bali, Indonesia Area. One of our client is already established in Surabaya, Indonesia and would like to extend it activity in Bali, Indonesia. They want our advice on where to locate their store.

<h2>A description of the data and how it will be used to solve the problem</h2>

We apply data analytics to help our client make better business decisions where to locate their store. We use region postal code data from https://kodeposresmi.com and crowd sourced data on platform like Foursquare.
